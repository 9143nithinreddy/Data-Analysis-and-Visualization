```python
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset: bill and tip
bill = np.array([34.81, 23.45, 44.30, 28.91, 35.42, 41.78, 20.00, 24.56, 18.43, 19.90])
tip = np.array([5.00, 3.50, 5.50, 4.00, 4.75, 5.75, 3.00, 3.75, 2.50, 3.00])

# Mean of bill and tip
mean_x = np.mean(bill)
mean_y = np.mean(tip)

# Total number of values
n = len(bill)

# Calculating the coefficients
numerator = np.sum((bill - mean_x) * (tip - mean_y))
denominator = np.sum((bill - mean_x) ** 2)
slope = numerator / denominator
intercept = mean_y - (slope * mean_x)

# Predicted tip values
predicted_tip = slope * bill + intercept

# Display the coefficients
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")

# Plotting the dataset and the regression line
plt.scatter(bill, tip, color='blue', label='Actual data')
plt.plot(bill, predicted_tip, color='red', label='Regression line')
plt.xlabel('Bill Amount')
plt.ylabel('Tip Amount')
plt.title('Bill vs Tip Linear Regression')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, img_as_float
from skimage.color import rgb2gray

# Load a sample image from skimage
image = img_as_float(data.astronaut())  # Convert image to float for processing

# Convert the image to grayscale
gray_image = rgb2gray(image)

# Invert the grayscale image
inverted_image = 1 - gray_image

# Adjust brightness by multiplying the grayscale image by a factor
brightness_factor = 1.5
brightened_image = np.clip(gray_image * brightness_factor, 0, 1)

# Display the original, grayscale, inverted, and brightened images
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
ax = axes.ravel()

ax[0].imshow(image)
ax[0].set_title("Original Image")
ax[0].axis('off')

ax[1].imshow(gray_image, cmap='gray')
ax[1].set_title("Grayscale Image")
ax[1].axis('off')

ax[2].imshow(inverted_image, cmap='gray')
ax[2].set_title("Inverted Image")
ax[2].axis('off')

ax[3].imshow(brightened_image, cmap='gray')
ax[3].set_title("Brightened Image")
ax[3].axis('off')

plt.show()
```
