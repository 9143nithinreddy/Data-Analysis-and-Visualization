```python
#a. Error Bars:
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(0, 10, 1)
y = np.sin(x)
yerr = 0.2  # Error values

# Plot with error bars
plt.errorbar(x, y, yerr=yerr, fmt='-o', ecolor='red', capsize=5, capthick=2)
plt.title("Error Bars")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#b. Boxplot with Whiskers:
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = np.random.normal(size=100)

# Boxplot
sns.boxplot(data=data)
plt.title("Boxplot with Whiskers")
plt.show()
```

```python
#c. Error Bars with Line plot:
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Sample data
x = np.arange(0, 10, 1)
y = np.cos(x)
yerr = 0.2  # Error values

# Line plot with error bars
plt.errorbar(x, y, yerr=yerr, fmt='-o', ecolor='green', capsize=5, capthick=2)
plt.title("Error Bars with Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#. Error Bars with Barplot
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 6, 3]
yerr = [0.5, 1.0, 0.7, 0.3]  # Error values

# Bar plot with error bars
plt.bar(categories, values, yerr=yerr, capsize=5, color='skyblue', edgecolor='black')
plt.title("Error Bars with Barplot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```
