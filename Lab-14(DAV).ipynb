```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'Apple-Fitness-data.csv'
fitness_data = pd.read_csv(file_path)

print(fitness_data.head())

fitness_data['DateTime'] = pd.to_datetime(fitness_data['Date'] + ' ' + fitness_data['Time'])

fitness_data.drop(['Date', 'Time'], axis=1, inplace=True)


fitness_data.dropna(inplace=True)

print(fitness_data.describe())

metrics = ['Step Count', 'Distance', 'Energy Burned', 'Flights Climbed', 
           'Walking Double Support Percentage', 'Walking Speed']

plt.figure(figsize=(12, 10))
for i, metric in enumerate(metrics, 1):
    plt.subplot(3, 2, i)
    sns.histplot(fitness_data[metric], kde=True)
    plt.title(f'Distribution of {metric}')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(x='DateTime', y='Step Count', data=fitness_data, label='Step Count')
sns.lineplot(x='DateTime', y='Distance', data=fitness_data, label='Distance')
sns.lineplot(x='DateTime', y='Energy Burned', data=fitness_data, label='Energy Burned')
plt.title('Time Series Analysis of Steps, Distance, and Energy Burned')
plt.legend()
plt.show()


plt.figure(figsize=(8, 6))
sns.heatmap(fitness_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Fitness Metrics')
plt.show()
```
