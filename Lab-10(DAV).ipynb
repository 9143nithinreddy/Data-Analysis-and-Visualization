```python
#1.1). Calculate coefficient of correlation between age of cars and annual maintenance and build scatter plot by showing correlation value on chart. 
```

```python
import numpy as np
import matplotlib.pyplot as plt
age_of_cars = np.array([2, 4, 6, 7, 8, 10, 12])
maintenance_cost = np.array([1600, 1500, 1800, 1900, 1700, 2100, 2000])
correlation_coefficient = np.corrcoef(age_of_cars, maintenance_cost)[0, 1]
plt.figure(figsize=(8, 6))
plt.scatter(age_of_cars, maintenance_cost, color='blue', label='Data Points')
plt.title('Scatter Plot: Age of Cars vs Annual Maintenance Cost')
plt.xlabel('Age of Cars (Years)')
plt.ylabel('Annual Maintenance Cost (Rupees)')
plt.text(3, 2050, f'Correlation: {correlation_coefficient:.2f}', fontsize=12, color='red')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#1.2). Demonstrate the different data visualization techniques. (Scatter Plot, Horizontal Bar Chart, Histogram, Visualization of Time Series data (Line Graphs) for applications such as weather analysis.(Weather nominal dataset).
```

```python
import pandas as pd

# Sample data for different visualizations
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 24, 36, 48, 52]
time_series_data = {
    'Date': pd.date_range(start='2023-01-01', periods=5, freq='D'),
    'Temperature': [25, 28, 23, 20, 22]
}

# Horizontal Bar Chart
plt.figure(figsize=(6, 4))
plt.barh(categories, values, color='purple')
plt.title('Horizontal Bar Chart')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.show()

# Histogram
plt.figure(figsize=(6, 4))
plt.hist(maintenance_cost, bins=4, color='orange')
plt.title('Histogram of Maintenance Costs')
plt.xlabel('Maintenance Cost (Rupees)')
plt.ylabel('Frequency')
plt.show()

# Line Graph (Time Series Data)
time_series_df = pd.DataFrame(time_series_data)
plt.figure(figsize=(8, 4))
plt.plot(time_series_df['Date'], time_series_df['Temperature'], marker='o', color='blue')
plt.title('Line Graph: Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
```

```python
#1.3). Generate Customizing Plot Legends, Customizing Color bars, Multiple Subplots on sample dataset.
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Sample Data for Subplots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Time series data
time_points = np.arange(1, 11)  # Time points from 1 to 10
time_series_data = [22, 24, 21, 25, 20, 23, 26, 24, 28, 27]  # Example temperature data

# Creating a figure with multiple subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# First subplot with legend customization
axs[0, 0].plot(x, y1, label='Sine Wave', color='blue')
axs[0, 0].plot(x, y2, label='Cosine Wave', color='orange')
axs[0, 0].set_title('Sine and Cosine Waves')
axs[0, 0].legend(loc='upper right', fontsize='small', title='Functions')

# Second subplot with a color bar
sc = axs[0, 1].scatter(x, y3, c=y3, cmap='viridis')
axs[0, 1].set_title('Tangent Function with Color Bar')
fig.colorbar(sc, ax=axs[0, 1], orientation='vertical', label='Tangent Values')

# Third subplot - simple line plot (Time Series Data)
axs[1, 0].plot(time_points, time_series_data, marker='o', linestyle='-', color='green')
axs[1, 0].set_title('Time Series Data')
axs[1, 0].set_xlabel('Time Points')
axs[1, 0].set_ylabel('Temperature (°C)')

# Fourth subplot - horizontal bar chart with custom colors
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 8, 4, 6]
axs[1, 1].barh(categories, values, color=cm.Paired(range(len(categories))))
axs[1, 1].set_title('Horizontal Bar Chart with Custom Colors')

plt.tight_layout()
plt.show()
```

```python
# 2.1)Generate density and contour plots using Python with Matplotlib and Seaborn.
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generating synthetic data
np.random.seed(0)
x = np.random.normal(size=1000)
y = np.random.normal(size=1000)

# Creating a density plot with seaborn
plt.figure(figsize=(12, 5))

# Density Plot
plt.subplot(1, 2, 1)
sns.kdeplot(x=x, y=y, fill=True, cmap='viridis')
plt.title('Density Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Contour Plot
plt.subplot(1, 2, 2)
sns.kdeplot(x=x, y=y, fill=False, cmap='viridis')
plt.title('Contour Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.tight_layout()
plt.show()
```
