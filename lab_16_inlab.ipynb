```python
import matplotlib.pyplot as plt

# Extracting the future preferences data
names = [student["Name"] for student in data]
future_preferences = [float(student["Future Preference"]) for student in data]

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.barh(names, future_preferences, color='skyblue')
plt.xlabel('Future Preference Score')
plt.ylabel('Student Name')
plt.title('Future Preferences of Students')
plt.show()
```

```python
import seaborn as sns

# Extracting department data
departments = [student["Department"] for student in data]

# Plotting the count of students per department
plt.figure(figsize=(8, 5))
sns.countplot(x=departments, palette='Set2')
plt.xlabel('Department')
plt.ylabel('Number of Students')
plt.title('Number of Students in Each Department')
plt.show()
```
