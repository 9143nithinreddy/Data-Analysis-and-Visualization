```python
import plotly.express as px

# Load the Gapminder dataset
df = px.data.gapminder()

# Remove rows with missing data
df_cleaned = df.dropna(subset=['lifeExp', 'gdpPercap', 'pop'])  # Clean dataset
```

```python
# Box plot for life expectancy by continent
box_plot = px.box(df_cleaned, x='continent', y='lifeExp', title='Life Expectancy by Continent (Box Plot)')
box_plot.show()

# Violin plot for life expectancy by continent
violin_plot = px.violin(df_cleaned, x='continent', y='lifeExp', title='Life Expectancy by Continent (Violin Plot)')
violin_plot.show()

# Interactive bar plot for average life expectancy by continent
bar_plot = px.bar(df_cleaned, x='continent', y='lifeExp', title='Average Life Expectancy by Continent', 
                  labels={'lifeExp':'Life Expectancy'}, 
                  color='continent', 
                  barmode='group')
bar_plot.show()
```

```python
# Normalize GDP per capita
df_cleaned['gdpPercap_normalized'] = (df_cleaned['gdpPercap'] - df_cleaned['gdpPercap'].min()) / \
                                      (df_cleaned['gdpPercap'].max() - df_cleaned['gdpPercap'].min())

# Scatter plot for lifeExp vs. normalized gdpPercap, colored by continent
scatter_plot = px.scatter(df_cleaned, x='lifeExp', y='gdpPercap_normalized', color='continent', 
                          title='Life Expectancy vs. Normalized GDP per Capita', 
                          labels={'gdpPercap_normalized':'Normalized GDP per Capita'})
scatter_plot.show()
```

```python
# Bubble chart with animation over time
timeline_animation = px.scatter(df_cleaned, x='lifeExp', y='gdpPercap', size='pop', color='country',
                                animation_frame='year', animation_group='country',
                                title='Timeline Animation: Life Expectancy vs GDP per Capita',
                                labels={'pop':'Population', 'gdpPercap':'GDP per Capita'},
                                hover_name='country', size_max=60)
timeline_animation.show()
```

```python
# 3D scatter plot for lifeExp, pop, and gdpPercap
scatter_3d = px.scatter_3d(df_cleaned, x='lifeExp', y='pop', z='gdpPercap', color='continent',
                           title='3D Plot: Life Expectancy, Population, and GDP per Capita',
                           labels={'pop':'Population', 'gdpPercap':'GDP per Capita', 'lifeExp':'Life Expectancy'})
scatter_3d.show()
```
